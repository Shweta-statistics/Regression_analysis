**To get more precise idea about multicollinearity we need to compute VIF(Varience Inflation Factor)**

```python
from sklearn.linear_model import LinearRegression

def VIF(data, feature):
    x_new = data.drop(columns=[feature])
    y_new = data[feature]
    model = LinearRegression()
    model.fit(x_new, y_new)
    y_pred = model.predict(x_new)
    sst = np.sum((y_new - np.mean(y_new))**2)
    sse = np.sum((y_new - y_pred)**2)
    r_squared = 1 - (sse / sst)
    if np.isclose(r_squared, 1):
        return np.inf

    return 1 / (1 - r_squared)
```

### **Handling Outliers Using the Hat Matrix in Regression**



```python
import numpy as np
import statsmodels.api as sm

X = np.asarray(sm.add_constant(x_n_2))
XtX_inv = np.linalg.pinv(X.T @ X)
H_diag = np.einsum('ij,ji->i', X @ XtX_inv, X.T)  # Efficient diagonal extraction

print(H_diag)
```

```python
p=X.shape[1]-1
n=X.shape[0]
threshold=(2*(p+1))/n
high_leverage_points=np.where(H_diag>threshold)[0]
print(high_leverage_points)
len(high_leverage_points)
```

### Now we have high levarage points but not all are influencial so we'll first use cooks distance to check for influencial points ###

```python
from statsmodels.stats.outliers_influence import OLSInfluence

model = sm.OLS(y, X).fit()
influence = OLSInfluence(model)
cooks_d = influence.cooks_distance[0]



cooks_threshold = 4 / n

influential_points = np.where(cooks_d > cooks_threshold)[0]


print("Influential points (indices):", influential_points)
```

```python
common_points = np.intersect1d(high_leverage_points, influential_points)
print("Points that are both high leverage and influential:", common_points)
len(common_points)
```

### So there are 11808 high leverage and influential points we'll remove them try to fit the new model ###

```python

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X_cleaned = np.delete(X, influential_points, axis=0)
y_cleaned = np.delete(y, influential_points, axis=0)

feature_names = ["Intercept"] + list(x_n_2.columns)
X_cleaned_df = pd.DataFrame(X_cleaned, columns=feature_names)

y_cleaned = y_cleaned.ravel()

model_cleaned = LinearRegression()
model_cleaned.fit(X_cleaned_df, y_cleaned)

y_pred_cleaned = model_cleaned.predict(X_cleaned_df)
r_squared_cleaned = r2_score(y_cleaned, y_pred_cleaned)

print("R-squared value is:", r_squared_cleaned)


```

```python
from numpy.linalg import pinv

from scipy.stats import chi2
from scipy.spatial.distance import mahalanobis

X_df = pd.DataFrame(X_cleaned)

# Compute Mahalanobis distance with pseudo-inverse or  Moore–Penrose inverse
cov_matrix = np.cov(X_df.values, rowvar=False)
inv_cov_matrix = np.linalg.pinv(cov_matrix)
mean_distr = X_df.mean(axis=0).values

mahalanobis_distances = X_df.apply(lambda row: mahalanobis(row.values, mean_distr, inv_cov_matrix), axis=1)

threshold = chi2.ppf(0.85, df=X_df.shape[1])
inliers = mahalanobis_distances < np.sqrt(threshold)

X_filtered = X_df[inliers]
y_filtered = y_cleaned[inliers]
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Create and fit the model
model = LinearRegression()
model.fit(X_filtered, y_filtered)

# Predict on the filtered data
y_pred = model.predict(X_filtered)

# Calculate R² score
r2 = r2_score(y_filtered, y_pred)
print(f"R² Score on filtered data: {r2:.4f}")
```

```python
import statsmodels.api as sm

def forward_selection(X, y):
    remaining = list(X.columns)
    selected = []
    current_score, best_new_score = -float('inf'), -float('inf')

    while remaining:
        scores_with_candidates = []
        for candidate in remaining:
            features = selected + [candidate]
            X_with_const = sm.add_constant(X[features], has_constant='add')
            model = sm.OLS(y, X_with_const).fit()
            score = model.rsquared_adj
            scores_with_candidates.append((score, candidate))

        scores_with_candidates.sort(reverse=True)
        best_new_score, best_candidate = scores_with_candidates[0]

        if best_new_score > current_score:
            remaining.remove(best_candidate)
            selected.append(best_candidate)
            current_score = best_new_score
        else:
            break

    X_final = sm.add_constant(X[selected], has_constant='add')
    final_model = sm.OLS(y, X_final).fit()
    return final_model, selected
```

```python
model, selected_features = forward_selection(X_check, y_check)
print(model.summary())
print("Selected features:", selected_features)

# Get residual sum of squares from forward-selected model
RSS_fs = np.sum(model.resid ** 2)

# Now fit the full model to get sigma^2
X_full_const = sm.add_constant(X_check, has_constant='add')
full_model = sm.OLS(y_check, X_full_const).fit()
RSS_full = np.sum(full_model.resid ** 2)

# Parameters
n = len(y_check)
p_fs = len(selected_features) + 1  # +1 for intercept
p_full = X_check.shape[1] + 1

# Estimate sigma^2 from full model
sigma_squared = RSS_full / (n - p_full)

# Mallows' Cp for the forward-selected model
Cp_fs = RSS_fs / sigma_squared - n + 2 * p_fs

print("Selected Features:", selected_features)
print("Mallows' Cp for Forward Selected Model:", Cp_fs)
```

```python
import statsmodels.api as sm

def backward_elimination(X, y, significance_level=0.05):
    features = list(X.columns)

    while True:
        X_with_const = sm.add_constant(X[features], has_constant='add')
        model = sm.OLS(y, X_with_const).fit()
        pvalues = model.pvalues.drop('const', errors='ignore')  # exclude intercept

        max_pval = pvalues.max()
        if max_pval > significance_level:
            worst_feature = pvalues.idxmax()
            features.remove(worst_feature)
        else:
            break

    final_model = sm.OLS(y, sm.add_constant(X[features], has_constant='add')).fit()
    return final_model, features
```
